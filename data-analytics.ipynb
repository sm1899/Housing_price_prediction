```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm, skew
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from typing import Dict, List, Tuple
from tqdm import tqdm
import warnings
import joblib
import os
import json
warnings.filterwarnings('ignore')

class ModelTuner:
    def __init__(self, X_train, X_val, y_train, y_val):
        self.X_train = X_train
        self.X_val = X_val
        self.y_train = y_train
        self.y_val = y_val
        
    def get_param_grids(self):
        """Define parameter grids for each model"""
        return {
            'Ridge': {
                'model': Ridge(),
                'params': {
                    'alpha': [0.1, 1.0, 5.0, 10.0, 20.0],
                    'solver': ['auto', 'svd', 'cholesky']
                }
            },
            'Lasso': {
                'model': Lasso(),
                'params': {
                    'alpha': [0.0001, 0.001, 0.01, 0.1, 1.0],
                    'selection': ['cyclic', 'random']
                }
            },
            'ElasticNet': {
                'model': ElasticNet(),
                'params': {
                    'alpha': [0.0001, 0.001, 0.01, 0.1, 1.0],
                    'l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9]
                }
            },
            'RandomForest': {
                'model': RandomForestRegressor(random_state=42),
                'params': {
                    'n_estimators': [100, 200],
                    'max_depth': [8, 10],
                    'min_samples_split': [2, 5],
                    'min_samples_leaf': [1, 2]
                }
            },
            'GradientBoosting': {
                'model': GradientBoostingRegressor(random_state=42),
                'params': {
                    'n_estimators': [1000, 2000],
                    'learning_rate': [0.01, 0.05],
                    'max_depth': [3, 4],
                    'min_samples_leaf': [10, 15],
                    'min_samples_split': [5, 10]
                }
            }
        }

    def evaluate_model(self, y_true, y_pred):
        """Calculate various regression metrics"""
        mse = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)
        
        return {
            'MSE': mse,
            'RMSE': rmse,
            'MAE': mae,
            'R2': r2
        }

    def tune_models(self):
        """Perform hyperparameter tuning for each model"""
        param_grids = self.get_param_grids()
        results = {}
        
        for name, config in tqdm(param_grids.items(), desc='Tuning models'):
            print(f"\nTuning {name}...")
            try:
                # Create GridSearchCV object
                grid_search = GridSearchCV(
                    estimator=config['model'],
                    param_grid=config['params'],
                    cv=5,
                    scoring='neg_mean_squared_error',
                    n_jobs=-1,
                    verbose=1
                )
                
                # Fit the grid search
                grid_search.fit(self.X_train, self.y_train)
                
                # Get best model
                best_model = grid_search.best_estimator_
                
                # Make predictions
                train_pred = best_model.predict(self.X_train)
                val_pred = best_model.predict(self.X_val)
                
                # Evaluate performance
                train_metrics = self.evaluate_model(self.y_train, train_pred)
                val_metrics = self.evaluate_model(self.y_val, val_pred)
                
                results[name] = {
                    'best_params': grid_search.best_params_,
                    'best_score': -grid_search.best_score_,  # Convert negative MSE back to positive
                    'train_metrics': train_metrics,
                    'val_metrics': val_metrics
                }
                
                # Print results
                print(f"\n{name} Results:")
                print(f"Best Parameters: {grid_search.best_params_}")
                print("\nValidation Metrics:")
                for metric, value in val_metrics.items():
                    print(f"{metric}: {value:.4f}")
                
            except Exception as e:
                print(f"Error tuning {name}: {str(e)}")
                results[name] = None
                
        return results

def run_tuning_pipeline(X_train, X_val, y_train, y_val):
    """Run the complete tuning pipeline"""
    tuner = ModelTuner(X_train, X_val, y_train, y_val)
    results = tuner.tune_models()
    
    # Print summary of best models
    print("\nModel Performance Summary:")
    print("-" * 50)
    for name, result in results.items():
        if result is not None:
            print(f"\n{name}:")
            print(f"Best Parameters: {result['best_params']}")
            print("\nValidation Metrics:")
            for metric, value in result['val_metrics'].items():
                print(f"{metric}: {value:.4f}")
    
    return results

class HousingEDA:
    def __init__(self, df: pd.DataFrame, target: str = 'SalePrice'):
        """
        Initialize the Housing EDA class.
        
        Args:
            df (pd.DataFrame): Input dataframe containing housing data
            target (str): Name of the target variable (default: 'SalePrice')
        """
        self.df = df.copy()  # Create a copy to avoid modifying original data
        self.target = target
        self.numeric_features = self.df.select_dtypes(include=['int64', 'float64']).columns
        self.categorical_features = self.df.select_dtypes(include=['object']).columns

    def analyze_feature_categories(self) -> Dict:
        """
        Analyze feature categories and their importance.
        
        Returns:
            Dict: Dictionary containing importance scores for each category
        """
        try:
            features_by_category = {
                'building': ['OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd',
                           'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond'],
                'space': ['GrLivArea', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF',
                         'GarageArea', 'GarageCars', 'FullBath', 'BedroomAbvGr'],
                'location': ['Neighborhood', 'MSZoning', 'LotArea', 'LotFrontage']
            }
            
            # Map ordinal features to numeric values
            ordinal_mapping = {
                'ExterQual': {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1},
                'ExterCond': {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1},
                'BsmtQual': {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0},
                'BsmtCond': {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0}
            }
            
            # Create a copy of the dataframe for analysis
            df_encoded = self.df.copy()
            
            # Apply ordinal mappings
            for feature, mapping in ordinal_mapping.items():
                if feature in df_encoded.columns:
                    df_encoded[feature] = df_encoded[feature].map(mapping).fillna(0)
            
            importance = {}
            for category, features in tqdm(features_by_category.items(), desc='Analyzing categories'):
                available_features = [f for f in features if f in df_encoded.columns]
                if not available_features:
                    continue
                    
                category_importance = []
                for feature in available_features:
                    if df_encoded[feature].dtype in ['int64', 'float64']:
                        # For numeric features, calculate correlation
                        correlation = df_encoded[feature].corr(df_encoded[self.target])
                        category_importance.append(abs(correlation))
                    else:
                        # For categorical features, calculate eta squared
                        categories = df_encoded[feature].unique()
                        grand_mean = df_encoded[self.target].mean()
                        n_total = len(df_encoded[self.target])
                        
                        ss_total = ((df_encoded[self.target] - grand_mean) ** 2).sum()
                        ss_between = sum(
                            len(df_encoded[df_encoded[feature] == cat]) * 
                            (df_encoded[df_encoded[feature] == cat][self.target].mean() - grand_mean) ** 2 
                            for cat in categories
                        )
                        
                        eta_squared = ss_between / ss_total if ss_total != 0 else 0
                        category_importance.append(eta_squared)
                
                if category_importance:
                    importance[category] = np.mean(category_importance)
            
            return importance
        except Exception as e:
            print(f"Error in feature category analysis: {str(e)}")
            return {}

    def analyze_missing_values(self) -> pd.DataFrame:
        """Analyze missing values in the dataset."""
        try:
            total = self.df.isnull().sum()
            percent = (self.df.isnull().sum() / len(self.df)) * 100
            missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
            missing_data = missing_data[missing_data['Total'] > 0]
            missing_data = missing_data.sort_values('Total', ascending=False)
            
            if len(missing_data) > 0:
                plt.figure(figsize=(10, 6))
                plt.bar(range(len(missing_data)), missing_data['Percent'])
                plt.xlabel('Features')
                plt.ylabel('Missing Percentage')
                plt.title('Missing Values Analysis')
                plt.xticks(range(len(missing_data)), missing_data.index, rotation=45)
                plt.tight_layout()
                plt.show()
            
            return missing_data
        except Exception as e:
            print(f"Error in missing values analysis: {str(e)}")
            return pd.DataFrame()

    def correlation_analysis(self, k: int = 10) -> pd.DataFrame:
        """Perform correlation analysis on numeric features."""
        try:
            numeric_df = self.df.select_dtypes(include=['int64', 'float64'])
            
            corrmat = numeric_df.corr()
            
            plt.figure(figsize=(12, 9))
            sns.heatmap(corrmat, vmax=.8, square=True)
            plt.title('Correlation Matrix')
            plt.tight_layout()
            plt.show()
        
            if self.target in numeric_df.columns:
                cols = corrmat.nlargest(k, self.target)[self.target].index
                cm = np.corrcoef(numeric_df[cols].values.T)
                
                plt.figure(figsize=(12, 9))
                sns.heatmap(cm, annot=True, square=True, fmt='.2f', 
                           xticklabels=cols, yticklabels=cols)
                plt.title(f'Top {k} Correlations')
                plt.tight_layout()
                plt.show()
            
            return corrmat
        except Exception as e:
            print(f"Error in correlation analysis: {str(e)}")
            return pd.DataFrame()

    def analyze_bivariate_relationships(self):
        """Analyze bivariate relationships between important features and target."""
        try:
            important_numerics = ['GrLivArea', 'TotalBsmtSF', 'OverallQual', 'YearBuilt']
            existing_features = [col for col in important_numerics if col in self.df.columns]
            
            for var in tqdm(existing_features, desc='Analyzing relationships'):
                plt.figure(figsize=(10, 6))
                plt.scatter(self.df[var], self.df[self.target], alpha=0.5)
                plt.xlabel(var)
                plt.ylabel(self.target)
                plt.title(f'{var} vs {self.target}')
                plt.tight_layout()
                plt.show()
        except Exception as e:
            print(f"Error in bivariate analysis: {str(e)}")

    def analyze_outliers(self):
        """Analyze outliers in the target variable and key features."""
        try:
            # Convert Series to numpy array and reshape for StandardScaler
            target_values = self.df[self.target].values.reshape(-1, 1)
            saleprice_scaled = StandardScaler().fit_transform(target_values)
            
            # Find outliers
            low_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][:10]
            high_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][-10:]
            
            print('Outer range (low) of the distribution:')
            print(low_range)
            print('\nOuter range (high) of the distribution:')
            print(high_range)
            
            if 'GrLivArea' in self.df.columns:
                plt.figure(figsize=(10, 6))
                plt.scatter(self.df['GrLivArea'], self.df[self.target], alpha=0.5)
                plt.xlabel('GrLivArea')
                plt.ylabel(self.target)
                plt.title('GrLivArea vs SalePrice - Outlier Detection')
                
                # Identify potential outliers using z-score
                z_scores = np.abs((self.df[self.target] - self.df[self.target].mean()) / self.df[self.target].std())
                outliers = self.df[z_scores > 3][[self.target, 'GrLivArea']]
                
                if len(outliers) > 0:
                    plt.scatter(outliers['GrLivArea'], outliers[self.target], 
                              color='red', alpha=0.7, label='Potential Outliers')
                    plt.legend()
                
                plt.show()
            
            return {
                'low_range': low_range.tolist(),
                'high_range': high_range.tolist(),
                'z_score_outliers_count': len(outliers) if 'outliers' in locals() else 0,
                'outlier_details': outliers.to_dict('records') if 'outliers' in locals() and len(outliers) > 0 else []
            }
        except Exception as e:
            print(f"Error in outlier analysis: {str(e)}")
            return {}

    def univariate_analysis(self):
        """Perform univariate analysis on the target variable."""
        try:
            plt.figure(figsize=(16, 6))
            
            plt.subplot(121)
            sns.distplot(self.df[self.target], fit=norm)
            plt.title(f'{self.target} Distribution')
            
            plt.subplot(122)
            stats.probplot(self.df[self.target], plot=plt)
            plt.title(f'{self.target} Probability Plot')
            
            plt.tight_layout()
            plt.show()
            
            print(f"Skewness: {self.df[self.target].skew()}")
            print(f"Kurtosis: {self.df[self.target].kurt()}")
        except Exception as e:
            print(f"Error in univariate analysis: {str(e)}")

    def test_assumptions(self):
        """Test statistical assumptions about the data."""
        try:
            self.univariate_analysis()
            
            for var in ['GrLivArea', 'TotalBsmtSF']:
                if var in self.df.columns:
                    plt.figure(figsize=(10, 6))
                    plt.scatter(self.df[var], self.df[self.target])
                    plt.xlabel(var)
                    plt.ylabel(self.target)
                    plt.title(f'Homoscedasticity Check: {var} vs {self.target}')
                    plt.tight_layout()
                    plt.show()
        except Exception as e:
            print(f"Error in testing assumptions: {str(e)}")

    def transform_features(self) -> pd.DataFrame:
        """Transform features to handle skewness and create new features."""
        try:
            df_transformed = self.df.copy()
            
            # Handle target skewness
            if df_transformed[self.target].skew() > 0.75:
                df_transformed[self.target] = np.log1p(df_transformed[self.target])
            
            # Transform GrLivArea if exists and skewed
            if 'GrLivArea' in df_transformed.columns and df_transformed['GrLivArea'].skew() > 0.75:
                df_transformed['GrLivArea'] = np.log1p(df_transformed['GrLivArea'])
            
            # Transform TotalBsmtSF if exists
            if 'TotalBsmtSF' in df_transformed.columns:
                df_transformed['HasBsmt'] = (df_transformed['TotalBsmtSF'] > 0).astype(int)
                mask = df_transformed['HasBsmt'] == 1
                df_transformed.loc[mask, 'TotalBsmtSF'] = np.log1p(df_transformed.loc[mask, 'TotalBsmtSF'])
            
            return df_transformed
        except Exception as e:
            print(f"Error in feature transformation: {str(e)}")
            return self.df

    def run_complete_analysis(self) -> Tuple[Dict, pd.DataFrame]:
        """Run complete EDA analysis and return results with transformed dataframe."""
        try:
            analysis_results = {}
            print("Starting comprehensive EDA...")
            
            # Basic statistics
            analysis_results['basic_stats'] = self.df.describe()
            
            # Missing values analysis
            print("\nAnalyzing missing values...")
            analysis_results['missing_data'] = self.analyze_missing_values()
            
            # Feature importance analysis
            print("\nAnalyzing feature categories...")
            analysis_results['feature_importance'] = self.analyze_feature_categories()
            
            # Correlation analysis
            print("\nAnalyzing correlations...")
            analysis_results['correlation_matrix'] = self.correlation_analysis()
            
            # Bivariate analysis
            print("\nAnalyzing bivariate relationships...")
            self.analyze_bivariate_relationships()
            
            # Outlier analysis
            print("\nAnalyzing outliers...")
            analysis_results['outliers'] = self.analyze_outliers()
            
            # Statistical assumptions
            print("\nTesting statistical assumptions...")
            self.test_assumptions()
            
            # Feature transformation
            print("\nTransforming features...")
            transformed_df = self.transform_features()
            
            return analysis_results, transformed_df
        
        except Exception as e:
            print(f"Error in complete analysis: {str(e)}")
            return {}, self.df

        

class HousingPipeline:
    def __init__(self, input_path: str = '../input/house-prices-advanced-regression-techniques/'):
        """
        Initialize Housing Pipeline.
        
        Args:
            input_path (str): Path to the input data directory
        """
        self.input_path = input_path
        self.train_path = os.path.join(input_path, 'train.csv')
        self.test_path = os.path.join(input_path, 'test.csv')
        self.model_path = './models/'
        os.makedirs(self.model_path, exist_ok=True)
        
        print("Loading data...")
        self.train = pd.read_csv(self.train_path)
        self.test = pd.read_csv(self.test_path)
        self.target = 'SalePrice'
        self.feature_names = None
        self.scaler = None
        self.best_model = None
    
    def run_eda(self):
        """Run exploratory data analysis."""
        try:
            eda = HousingEDA(self.train)
            results, self.train = eda.run_complete_analysis()
            return results
        except Exception as e:
            print(f"Error in EDA: {str(e)}")
            raise
    
    def handle_missing_values(self):
        """Handle missing values in both train and test datasets."""
        print("\nHandling missing values...")
        try:
            all_data = pd.concat([self.train, self.test], axis=0, sort=False)
            
            for col in tqdm(all_data.columns, desc='Processing columns'):
                if all_data[col].isnull().sum() > 0:
                    if all_data[col].dtype in ['int64', 'float64']:
                        all_data[col] = all_data[col].fillna(all_data[col].mean())
                    else:
                        all_data[col] = all_data[col].fillna('Missing')
            
            train_len = len(self.train)
            self.train = all_data[:train_len]
            self.test = all_data[train_len:]
        except Exception as e:
            print(f"Error in handling missing values: {str(e)}")
            raise
    
    def feature_engineering(self):
        """Perform feature engineering on both train and test datasets."""
        print("\nPerforming feature engineering...")
        try:
            all_data = pd.concat([self.train, self.test], axis=0, sort=False)
            
            # Create new features
            feature_engineering_steps = [
                ('TotalSF', lambda x: x['TotalBsmtSF'] + x['1stFlrSF'] + x['2ndFlrSF']),
                ('TotalBath', lambda x: x['FullBath'] + 0.5 * x['HalfBath'] + 
                 x['BsmtFullBath'] + 0.5 * x['BsmtHalfBath']),
                ('HasPool', lambda x: (x['PoolArea'] > 0).astype(int)),
                ('Has2ndFloor', lambda x: (x['2ndFlrSF'] > 0).astype(int)),
                ('HasGarage', lambda x: (x['GarageArea'] > 0).astype(int)),
                ('HasBsmt', lambda x: (x['TotalBsmtSF'] > 0).astype(int)),
                ('Remodeled', lambda x: (x['YearRemodAdd'] != x['YearBuilt']).astype(int)),
                ('Age', lambda x: 2024 - x['YearBuilt']),
                ('TimeSinceRemodel', lambda x: 2024 - x['YearRemodAdd'])
            ]
            
            for name, operation in tqdm(feature_engineering_steps, desc='Creating features'):
                try:
                    all_data[name] = operation(all_data)
                except Exception as e:
                    print(f"Error creating feature {name}: {str(e)}")
            
            # Log transform skewed numeric features
            numeric_feats = all_data.dtypes[all_data.dtypes != "object"].index
            skewed_feats = all_data[numeric_feats].apply(lambda x: skew(x.dropna()))
            skewed_feats = skewed_feats[abs(skewed_feats) > 0.75]
            
            for feat in tqdm(skewed_feats.index, desc='Transforming skewed features'):
                all_data[feat] = np.log1p(all_data[feat])
            
            # Convert categorical variables to dummy variables
            all_data = pd.get_dummies(all_data)
            
            train_len = len(self.train)
            self.train = all_data[:train_len]
            self.test = all_data[train_len:]
            
        except Exception as e:
            print(f"Error in feature engineering: {str(e)}")
            raise
    
    def prepare_data(self):
        """Prepare data for modeling by scaling features and splitting train/validation sets."""
        print("\nPreparing data for modeling...")
        try:
            # Split features and target
            X = self.train.drop(self.target, axis=1)
            y = self.train[self.target]
            
            # Store feature names
            self.feature_names = X.columns.tolist()
            
            # Scale features
            self.scaler = RobustScaler()
            X_scaled = self.scaler.fit_transform(X)
            X_scaled = pd.DataFrame(X_scaled, columns=self.feature_names)
            
            return train_test_split(X_scaled, y, test_size=0.2, random_state=42)
            
        except Exception as e:
            print(f"Error in data preparation: {str(e)}")
            raise
    
    def get_models(self):
        """Get dictionary of models to train."""
        return {
            'Linear': LinearRegression(),
            'Ridge': Ridge(alpha=1.0),
            'Lasso': Lasso(alpha=0.001),
            'ElasticNet': ElasticNet(alpha=0.001, l1_ratio=0.5),
            'SVR': SVR(kernel='rbf'),
            'RandomForest': RandomForestRegressor(
                n_estimators=300,
                max_depth=10,
                min_samples_split=5,
                min_samples_leaf=2,
                random_state=42
            ),
            'GradientBoosting': GradientBoostingRegressor(
                n_estimators=3000,
                learning_rate=0.05,
                max_depth=4,
                max_features='sqrt',
                min_samples_leaf=15,
                min_samples_split=10,
                random_state=42
            )
        }

    def train_models(self, X_train, X_val, y_train, y_val):
        """Train multiple models and evaluate their performance."""
        print("\nTraining models...")
        models = self.get_models()
        results = {}
        
        for name, model in tqdm(models.items(), desc='Training models'):
            try:
                # Training
                model.fit(X_train, y_train)
                
                # Predictions
                train_pred = model.predict(X_train)
                val_pred = model.predict(X_val)
                
                # Cross-validation
                cv_scores = cross_val_score(
                    model, X_train, y_train,
                    cv=5,
                    scoring='neg_mean_squared_error'
                )
                rmse_cv = np.sqrt(-cv_scores)
                
                # Calculate metrics
                results[name] = {
                    'train_rmse': np.sqrt(mean_squared_error(y_train, train_pred)),
                    'val_rmse': np.sqrt(mean_squared_error(y_val, val_pred)),
                    'cv_rmse_mean': rmse_cv.mean(),
                    'cv_rmse_std': rmse_cv.std()
                }
                
                # Save model
                model_path = os.path.join(self.model_path, f'{name.lower()}_model.joblib')
                joblib.dump(model, model_path)
                
                # Save feature importances if available
                if hasattr(model, 'feature_importances_'):
                    importances = pd.DataFrame({
                        'feature': self.feature_names,
                        'importance': model.feature_importances_
                    })
                    importances = importances.sort_values('importance', ascending=False)
                    importances.to_csv(
                        os.path.join(self.model_path, f'{name.lower()}_importances.csv')
                    )
                
            except Exception as e:
                print(f"Error training {name} model: {str(e)}")
                results[name] = None
        
        # Save results
        with open(os.path.join(self.model_path, 'model_results.json'), 'w') as f:
            json.dump(results, f, indent=4)
        
        return results, models

    def transform_test_data(self):
        """Transform test data using the fitted scaler."""
        try:
            # Ensure test data has same columns as training data
            missing_cols = set(self.feature_names) - set(self.test.columns)
            extra_cols = set(self.test.columns) - set(self.feature_names)
            
            # Add missing columns with 0s
            for col in missing_cols:
                self.test[col] = 0
            
            # Remove extra columns
            self.test = self.test.drop(columns=list(extra_cols))
            
            # Ensure column order matches training data
            self.test = self.test[self.feature_names]
            
            # Transform using scaler
            X_test_scaled = self.scaler.transform(self.test)
            return X_test_scaled
            
        except Exception as e:
            print(f"Error in test data transformation: {str(e)}")
            raise

    def make_predictions(self):
        """Make predictions on test data using best model."""
        try:
            # Transform test data
            X_test_scaled = self.transform_test_data()
            
            # Make predictions
            predictions = np.expm1(self.best_model.predict(X_test_scaled))
            
            # Create submission file
            submission = pd.DataFrame({
                'Id': self.test.index,
                'SalePrice': predictions
            })
            submission.to_csv('submission.csv', index=False)
            
            return submission
            
        except Exception as e:
            print(f"Error in making predictions: {str(e)}")
            raise

    def run_pipeline(self):
        """Execute complete pipeline from data loading to model training."""
        try:
            # Run EDA
            print("Running EDA...")
            eda_results = self.run_eda()
            
            # Data preprocessing
            self.handle_missing_values()
            self.feature_engineering()
            
            # Prepare data
            X_train, X_val, y_train, y_val = self.prepare_data()
            
            # Add hyperparameter tuning
            tuning_results = run_tuning_pipeline(X_train, X_val, y_train, y_val)
            
            # Select best model based on validation RMSE
            best_model_name = min(
                tuning_results,
                key=lambda k: tuning_results[k]['val_metrics']['RMSE'] if tuning_results[k] is not None else float('inf')
            )
            
            # Get the best model configuration
            best_params = tuning_results[best_model_name]['best_params']
            
            # Initialize and train the best model with optimal parameters
            models = self.get_models()
            self.best_model = models[best_model_name].set_params(**best_params)
            self.best_model.fit(X_train, y_train)
            
            print(f"\nBest Model: {best_model_name}")
            print(f"Best Parameters: {best_params}")
            print("\nBest Model Validation Metrics:")
            for metric, value in tuning_results[best_model_name]['val_metrics'].items():
                print(f"{metric}: {value:.4f}")
            
            return tuning_results, self.best_model
            
        except Exception as e:
            print(f"Error in pipeline execution: {str(e)}")
            raise


if __name__ == "__main__":
    try:
        # Initialize and run pipeline
        pipeline = HousingPipeline()
        
        # Run complete pipeline
        tuning_results, best_model = pipeline.run_pipeline()
        
        # Make predictions and create submission file
        submission = pipeline.make_predictions()
        
        print("\nPipeline complete! Check 'submission.csv' for predictions.")
        
    except Exception as e:
        print(f"Pipeline failed: {str(e)}")
```
